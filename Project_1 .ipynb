```python
df.head()
```

# Gender distribution

```python
plt.figure(figsize=(4,5))
ax = sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.plot()
plt.show()
```

```python
#From the above chart we have analysed that the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

# Representing data Heatmap

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship Between Parent's Education and Student's Score", color='Brown', size=14)
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship Between Parent's Marital Status and Student's Score",color='Brown', size=14)
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the 
#student's score due to parents marital status
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]

mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups", color='Brown', size=14)
plt.show()
```
